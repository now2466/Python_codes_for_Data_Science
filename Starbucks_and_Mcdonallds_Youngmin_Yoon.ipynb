```python
# 두 좌표간의 거리를 구하기 위해서 haversine 모듈을 사용했습니다
from haversine import haversine

num = 0

for idx,rows in all_data.iterrows():
    start = (rows["lat"],rows["lng"])
    flag=True
    i=1
    #flag가 false가 되기 전까지 계속 반복
    
    while(i<len(station_data)):
        end = (station_data["lat"][i],station_data["lng"][i])
        dist=haversine(start,end) #거리계산. 디폴트가 1km단위이기 때문에 0.1이면 100m
        if dist<0.1: # 가게와 주유소간의 거리가 100미터 이내인 곳을 찾으면 flag가 falsefh 바뀌고 while문이 종료된다.
            num+=1
            break
        else:
            i+=1 # i를 증가시켜주어 다음 주유소를 확인
num
```
